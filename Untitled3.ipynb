```python
import os
import mne

# Define the download directory
download_dir = './sleep-edfx_downloaded/'
os.makedirs(download_dir, exist_ok=True)

# Download the dataset using wget
# -r: recursive download
# -N: timestamping (don't re-retrieve files unless newer)
# -c: continue a broken download
# -np: no parent directory (don't ascend to the parent directory)
print(f"Downloading dataset to {download_dir}...")
!wget -r -N -c -np -P {download_dir} https://physionet.org/files/sleep-edfx/1.0.0/

# Construct the file paths for loading after download
# Assuming the structure after download is ./sleep-edfx_downloaded/physionet.org/files/sleep-edfx/1.0.0/sleep-cassette/SC-subjects/SC-recordings/
base_path = os.path.join(download_dir, 'physionet.org', 'files', 'sleep-edfx', '1.0.0', 'sleep-cassette')

subjects = [0, 1]
recordings = [1]
paths = []
for subject in subjects:
    for recording in recordings:
        subject_str = f'SC{subject+1:02d}'
        recording_str = f'SC{subject+1:02d}E{recording}'
        filename = f'{recording_str}-PSG.edf'
        file_path = os.path.join(base_path, subject_str, recording_str, filename)
        paths.append(file_path)

# Load the EDF files
raw_objects = []
for path in paths:
    try:
        raw = mne.io.read_raw_edf(path, preload=True, verbose=False) # Set verbose to False to reduce output
        raw_objects.append(raw)
        print(f"Successfully loaded {os.path.basename(path)}")
    except FileNotFoundError:
        print(f"File not found after download: {path}. Skipping.")
    except Exception as e:
        print(f"Error loading file {path}: {e}. Skipping.")

# Print a confirmation message
if raw_objects:
    print("\nSleep-EDF Expanded dataset downloaded and loaded successfully.")
else:
    print("\nFailed to download and load any Sleep-EDF Expanded dataset files.")
```
